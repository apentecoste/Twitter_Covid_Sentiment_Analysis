```python
import pandas as pd
import numpy as np
import GetOldTweets3 as got
```

```python
def text_query_to_csv(text_query,  since, until,state,lang):
    
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setSince(since).setUntil(until).setNear(state).setLang(lang)
                                                    
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)

    text_tweets = [[tweet.date, tweet.text, tweet.username, tweet.retweets, tweet.mentions, tweet.to, tweet.permalink] for tweet in tweets]

    tweets_df = pd.DataFrame(text_tweets, columns = ['Datetime', 'Text', 'Username','Retweets','Mentions','Reply','Link'])
    #tweets_df.to_csv(str(since)+"_"+str(state)+"_"+str(until)+"-"+str(lang)+"-tweets.csv", sep=',')
    return(tweets_df)
```
